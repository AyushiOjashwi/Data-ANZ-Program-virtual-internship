# TRANSACTION VOLUME EACH DAY

```python
np1_group = np1.groupby('date')["amount"].sum().reset_index()
np1_group.head()
```

```python
np1_date_count = np1.groupby("date").count()
trans_vol = np1_date_count["customer_id"].mean()
n_points = len(np1_date_count.index)

plt.figure(figsize=(20, 7))
plt.plot(np1_date_count.index, np1_date_count["customer_id"], c="green", label = "Customer ID")
plt.plot(np1_date_count.index, numpy.linspace(trans_vol,trans_vol,n_points), c="r", label = "Mean transaction volume")
plt.title("ANZ Transaction Volume vs. Date")
plt.xlabel("Date")
plt.ylabel("Number of customers")
plt.legend(loc='lower center')
plt.tight_layout()
```

# Mean transaction amount each day

```python
np1_date_mean = np1.groupby("date").mean()
trans_amt = np1_date_mean["amount"].mean()
n_points = len(np1_date_count.index)
plt.figure(figsize=(20,7))
plt.plot(np1_date_count.index, np1_date_mean["amount"], c="grey", label = "Amount")
plt.plot(np1_date_count.index, numpy.linspace(trans_amt,trans_amt,n_points), c="r", label = "Overall mean transaction amount") 
plt.title("ANZ Mean Transaction Amount vs. Date")
plt.xlabel("Date")
plt.ylabel("Amount ($)")
plt.legend()
plt.tight_layout()
```

```python
months = []
for date in np1["date"]:
    if date.month == 8:
        months.append("August")
    elif date.month == 9:
        months.append("September")
    elif date.month == 10:
        months.append("October")

np1["Months"] = months
np1["Months"].head()
#code by Ayushi
```

# Mean customer balance and payment amount by age

```python
np1_cus_aug = np1[np1["Months"] == "August"].groupby("customer_id").mean()
np1_gen_aug = np1[np1["Months"] == "August"].groupby("gender").mean()

mean_f_bal_aug = np1_gen_aug["balance"].iloc[0]
mean_m_bal_aug = np1_gen_aug["balance"].iloc[1]
n_points = len(np1_cus_aug["age"])
```

```python
x = np1['age']
plt.xlabel("Age (years)")
plt.ylabel("number of customers")
num_bins = 5
n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
plt.show()
#no of people in the following age group are shown in the below graph.
```

```python
plt.scatter(np1_cus_aug["age"], np1_cus_aug["balance"], c="green", label="Balance")
plt.plot(np1_cus_aug["age"], numpy.linspace(mean_f_bal_aug,mean_f_bal_aug,n_points), c="r", label = "Mean female balance")
plt.plot(np1_cus_aug["age"], numpy.linspace(mean_m_bal_aug,mean_m_bal_aug,n_points), c="purple", label = "Mean male balance")
plt.title("ANZ Customer Balance vs. Age for August")
plt.xlabel("Age (years)")
plt.ylabel("Balance ($)")
plt.legend()
plt.tight_layout()
```

```python
mean_f_amt_aug = np1_gen_aug["amount"].iloc[0]
mean_m_amt_aug = np1_gen_aug["amount"].iloc[1]
plt.scatter(np1_cus_aug["age"], np1_cus_aug["amount"], c="pink", label="Amount")
plt.plot(np1_cus_aug["age"], numpy.linspace(mean_f_amt_aug,mean_f_amt_aug,n_points), c="r", label = "Mean female amount")
plt.plot(np1_cus_aug["age"], numpy.linspace(mean_m_amt_aug,mean_m_amt_aug,n_points), c="grey", label = "Mean male amount")
plt.title("ANZ Customer Mean Payment Amount vs. Age for August")
plt.xlabel("Age (years)")
plt.ylabel("Amount ($)")
plt.legend()
plt.tight_layout()
```
